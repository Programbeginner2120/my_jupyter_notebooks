```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_data = pd.read_csv("./datasets/spotify_songs.csv")
spotify_data.head()
```

```python
spotify_data.dropna(inplace=True)
spotify_data.isna().sum()
```

```python
spotify_data.columns
```

```python
fig, ax = plt.subplots()
ax.bar(spotify_data['playlist_genre'], spotify_data['track_popularity'])

ax.set_xlabel("Playlist Genre")
ax.set_ylabel("Track Popularity")
ax.set_title("Playlist Genre vs. Track Popularity")

plt.tight_layout()
plt.show()
```

```python
tracks_to_omit = ['track_id', 'track_name', 'track_artist', 'track_album_id', 'track_album_name',
                  'track_album_release_date', 'playlist_name', 'playlist_id',]
spotify_data_updated = spotify_data.drop(columns=tracks_to_omit)
spotify_data_updated.columns
```

```python
plt.bar(spotify_data_updated['playlist_subgenre'], spotify_data_updated['track_popularity'], width=0.6)

plt.xticks(rotation=90)

plt.xlabel("Playlist Subgenre")
plt.ylabel("Track Popularity")
plt.title("Playlist Subgenre vs. Track Popularity")


plt.tight_layout()


plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
popularity = spotify_data_updated['track_popularity']
X = spotify_data_updated.drop(columns=['track_popularity']).select_dtypes(include='number')

# print(popularity.shape)
# print(X.shape)

X_train, X_test, y_train, y_test = train_test_split(X, popularity, test_size=0.2)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(X_test.shape[0], y_test.shape, y_pred.shape)

mse = mean_squared_error(y_test, y_pred)
print(f"Mean squared error: {mse}")

# Plot the regression line
plt.scatter(np.arange(X_test.shape[0])[::50], y_test[::50], color='red')
plt.scatter(np.arange(X_test.shape[0])[::50], y_pred[::50], color='blue')
plt.title('Linear Regression Model')
plt.show()
```
