```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
spotify_data = pd.read_csv("../datasets/spotify_songs.csv")
spotify_data.head()
```

```python
spotify_data.dropna(inplace=True)
spotify_data.isna().sum()
```

```python
spotify_data.columns
```

```python
fig, ax = plt.subplots()
average_popularity = spotify_data.groupby('playlist_genre')['track_popularity'].mean()
ax.bar(average_popularity.index, average_popularity.values)

ax.set_xlabel("Playlist Genre")
ax.set_ylabel("Track Popularity")
ax.set_title("Playlist Genre vs. Track Popularity")

min_val, max_val = average_popularity.min(), average_popularity.max()
padding = (max_val - min_val) * 0.1
ax.set_ylim(min_val - padding, max_val + padding)

plt.tight_layout()
plt.show()
```

```python
tracks_to_omit = ['track_id', 'track_name', 'track_artist', 'track_album_id', 'track_album_name',
                  'track_album_release_date', 'playlist_name', 'playlist_id',]
spotify_data_updated = spotify_data.drop(columns=tracks_to_omit)
spotify_data_updated.columns
```

```python
average_sub_popularity = spotify_data_updated.groupby('playlist_subgenre')['track_popularity'].mean()

plt.bar(average_sub_popularity.index, average_sub_popularity.values, width=0.6)

plt.xticks(rotation=90)

plt.xlabel("Playlist Subgenre")
plt.ylabel("Track Popularity")
plt.title("Playlist Subgenre vs. Track Popularity")

min_val, max_val = average_sub_popularity.min(), average_sub_popularity.max()
padding = (max_val - min_val) * 0.1
plt.ylim(min_val - padding, max_val + padding)

plt.tight_layout()

plt.show()
```

Let's see the correlations between variables in a heatmap data structure

```python
correlation_matrix = spotify_data_updated.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, fmt=".2f", annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

We can see a few things here:
1. Energy is highly positively correlated with loudness (0.68)
2. Energy is moderately negatively correlated with acousticness (-0.54)
3. Loudness is moderately negatively correlated with acousticness (-0.36)
4. Valence is moderately positively correlated with danceability (0.33)
5. Track popularity is lowly negatively correlated with instrumentalness (-0.15), duration_ms (-0.14) and energy (-0.11) 
5. There are low positive and negative correlations with other variables

Let's see the relationship between duration ms and track popularity

```python
from scipy.stats import linregress

spotify_data_updated['duration_seconds'] = spotify_data_updated['duration_ms']  / 1000
spotify_data_updated.drop(columns=['duration_ms'], inplace=True)

slope, intercept, _, _, _ = linregress(spotify_data_updated['duration_seconds'], spotify_data_updated['track_popularity'])
equation = f"y = {slope:.2f}x + {intercept:.2f}"

sns.regplot(data=spotify_data_updated, x='duration_seconds', y='track_popularity', line_kws={'color': 'red'})
plt.title("Duration in Seconds Vs. Track Popularity")

plt.text(0.1, 98.5, equation)
```

We can see that this data is described by a negatively sloped linear regression line, with a slope of -0.6 and an intercept of 56.02
