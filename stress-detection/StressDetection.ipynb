Let's load the dataset with headers and see some basics about it

Let's take a look at the correlation between variables

```python
correlation_matrix = df.drop(columns=['participant_id', 'day']).corr()

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, fmt=".2f", annot=True, cmap="coolwarm", vmin=1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

It's interesting to see that our variables are not very correlated with one another

Let's take a look at some of the variables and how they vary by day. Let's take a look at PSS score and PSQI score.

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 14))

sns.barplot(x='day', y='PSS_score', data=df, ax=axes[0])
axes[0].set_title('Day vs. PSS_score')
sns.barplot(x='day', y='PSQI_score', data=df, ax=axes[1])
axes[1].set_title('Day vs. PSQI_score')

plt.tight_layout()
plt.show()
```
