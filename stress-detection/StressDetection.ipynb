Let's take a look at the correlation between variables

Let's take a look at some of the variables and how they vary by day. Let's take a look at PSS score and PSQI score.

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 14))

sns.barplot(x='day', y='PSS_score', data=df, ax=axes[0], errorbar=None)
axes[0].set_title('Day vs. PSS_score')
min_val, max_val = df.groupby('day')['PSS_score'].mean().min(), df.groupby('day')['PSS_score'].mean().max()
padding = (max_val - min_val) * .1
axes[0].set_ylim(min_val - padding, max_val + padding)

sns.barplot(x='day', y='PSQI_score', data=df, ax=axes[1], errorbar=None)
axes[1].set_title('Day vs. PSQI_score')
min_val, max_val = df.groupby('day')['PSQI_score'].mean().min(), df.groupby('day')['PSQI_score'].mean().max()
padding = (max_val - min_val) * .1
axes[1].set_ylim(min_val - padding, max_val + padding)

plt.tight_layout()
plt.show()
```

We can see that the PSS_score and the PSQI_score vary greatly by day on average. Let's try to build a model to predict each one respectively.

```python
from sklearn.linear_model import LinearRegression

df_without_pss_score_and_psqi_score = df.drop(columns=['PSS_score', 'PSQI_score'])

lr_pss_score = LinearRegression()
X = df_without_pss_score_and_psqi_score
Y_pss_score = df['PSS_score']
lr_pss_score.fit(X, Y_pss_score)

print(f"COEFFICIENT: {lr_pss_score.coef_}")
print(f"INTERCEPT: {lr_pss_score.intercept_}")

lr_psqi_score = LinearRegression()
Y_psqi_score = df['PSQI_score']
lr_psqi_score.fit(X, Y_psqi_score)

print(f"COEFFICIENT: {lr_psqi_score.coef_}")
print(f"INTERCEPT: {lr_psqi_score.intercept_}")
```

Let's see how this model does on predictions and see the difference betweent the predicted vs. actual values

```python
from sklearn.metrics import mean_squared_error

y_hat_pss_score = lr_pss_score.predict(X)
mse_pss_score = mean_squared_error(Y_pss_score, y_hat_pss_score)
print(f"Mean Squared Error PSS_score: {mse_pss_score}")

y_hat_psqi_score = lr_psqi_score.predict(X)
mse_psqi_score = mean_squared_error(Y_psqi_score, y_hat_psqi_score)
print(f"Mean Squared Error PSQI_score: {mse_psqi_score}")
```

Let's see if we can improve these metrics by using cross validation. Furthermore, let's factor in feature scaling and pipelines to make our regression more effective and our lives easier.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import r2_score

# Splitting up our datasets into training and testing sets
X_train_val_pss, X_test_pss, y_train_val_pss, y_test_pss = train_test_split(X, Y_pss_score, test_size=0.2, random_state=42)
X_train_val_psqi, X_test_psqi, y_train_val_psqi, y_test_psqi = train_test_split(X, Y_psqi_score, test_size=0.2, random_state=42)

# defining our pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

# scoring our models
pss_cross_val_scores = cross_val_score(pipeline, X_train_val_pss, y_train_val_pss, cv=5, scoring='r2')

print(f"PSS Cross-Validation R^2 Scores: {pss_cross_val_scores}")
print(f"PSS Mean R^2 Score: {np.mean(pss_cross_val_scores)}")

psqi_cross_val_scores = cross_val_score(pipeline, X_train_val_psqi, y_train_val_psqi, cv=5, scoring='r2')

print(f"PSQI Cross-Validation R^2 Scores: {psqi_cross_val_scores}")
print(f"PSQI Mean R^2 Score: {np.mean(psqi_cross_val_scores)}")
```

These models don't look very promising, but in any case let's generate some predictions with these models

```python
pss_predictions = cross_val_predict(pipeline, X_train_val_pss, y_train_val_pss, cv=5)
psqi_predictions = cross_val_predict(pipeline, X_train_val_psqi, y_train_val_psqi, cv=5)

mse_cv_pss_score = mean_squared_error(y_train_val_pss, pss_predictions)
mse_cv_psqi_score = mean_squared_error(y_train_val_psqi, psqi_predictions)

print(f"PSS MSE: {mse_cv_pss_score}")
print(f"PSQI MSE: {mse_cv_psqi_score}")
```
